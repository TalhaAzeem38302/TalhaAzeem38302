```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_data['WindGustDir'] = le.fit_transform(df_data['WindGustDir'])
df_data['WindDir9am'] = le.fit_transform(df_data['WindDir9am'])
df_data['WindDir3pm'] = le.fit_transform(df_data['WindDir3pm'])
df_data['RainToday'] = le.fit_transform(df_data['RainToday'])
df_data['RainTomorrow'] = le.fit_transform(df_data['RainTomorrow'])
```

```python
x = df_data.drop(['RainTomorrow'], axis = 1)
y = df_data['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
predictions = lr.predict(x_test)
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
print(accuracy_score(y_test, predictions))
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
predictions = dt.predict(x_test)
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
print(accuracy_score(y_test, predictions))
```

```python
#Random Forest Classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
predictions = rf.predict(x_test)
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
print(accuracy_score(y_test, predictions))
```

```python
#XGBoost Classifier
```

```python
import xgboost as xgb
xgb = xgb.XGBClassifier()
xgb.fit(x_train, y_train)
pred = xgb.predict(x_test)
print('acc',accuracy_score(y_test,pred))
print('f1',classification_report(y_test,pred))
print('matrix',confusion_matrix(y_test,pred))
```
